```python
import torch
from torch import nn
import torch.nn.functional as F
```

```python
def attention(Q,K,V):
    # Q batch,seq_len,embedding_size
    batch_size,seql_len,embed_size = Q.shape
    _,tmp,_ = K.shape
    weights = Q @ K.transpose(-1,-2) 
    mask = torch.tril(torch.ones((seql_len,tmp))) 
    weights = torch.masked_fill(weights,mask==0.,-torch.inf)
#     print(weights)
    weights = F.softmax(weights,dim=-1)
#     print(weights)
    V = weights @ V
    return V
```

```python
class MultiHead(nn.Module):
    
    def __init__(self,q_size,k_size,v_size,num_heads,embedding_size):
        super().__init__()
        self.num_heads = num_heads
        self.q_linear = nn.Linear(q_size,embedding_size//num_heads)
        self.k_linear = nn.Linear(k_size,embedding_size//num_heads)
        self.v_linear = nn.Linear(v_size,embedding_size//num_heads)
    
    
    def forward(self,Q,K,V):
        queries = self.q_linear(Q)
        keys = self.k_linear(K)
        values = self.v_linear(V)
        ans = [attention(queries,keys,values) for _ in range(self.num_heads)]
        return torch.cat(ans,axis=-1)        
```

```python
class FFN(nn.Module):
    
    def __init__(self,embed_size,hidden_size,output_size):
        super().__init__()
        self.linear1 = nn.Linear(embed_size,hidden_size)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(hidden_size,output_size)
    
    def forward(self,x):
        x = self.linear1(x)
        x = self.relu(x)
        return self.linear2(x)
```

```python
class Block(nn.Module):
    
    def __init__(self,q_size,k_size,v_size,num_heads,embedding_size,hidden_size):
        super().__init__()
        self.mh = MultiHead(q_size,k_size,v_size,num_heads,embedding_size)
        self.ln1 = nn.LayerNorm(embedding_size)
        self.f1 = FFN(embedding_size,hidden_size,embedding_size)
        self.ln2 = nn.LayerNorm(embedding_size)
    
    def forward(self,Q,K,V):
        t = self.mh(Q,K,V)
        x = V + t
        x = self.ln1(x)
        x = x + self.f1(x)
        x = self.ln2(x)
        return x
```

```python
Q = torch.ones((2,3,100))
K = torch.ones((2,3,100))
V = torch.ones((2,3,100))
```

```python
mh = MultiHead(Q.shape[-1],K.shape[-1],V.shape[-1],5,100)
```

```python
class Encoder(nn.Module):
    
    def __init__(self,q_size,k_size,v_size,num_heads,embedding_size,hidden_size,block_size,vocab_size):
        super().__init__()
        self.emdedding = nn.Embedding(vocab_size,embedding_size)
        self.blks = [Block(q_size,k_size,v_size,num_heads,embedding_size,hidden_size) for _ in range(block_size)]
        self.dense = nn.Linear(embedding_size,vocab_size)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self,x):
        for blk in self.blks:
            x = blk(x,x,x)
        return x
```

```python
class DecoderBlock(nn.Module):
    
    def __init__(self,q_size,k_size,v_size,num_heads,embedding_size,hidden_size):
        super().__init__()
        self.mh1 = MultiHead(q_size,k_size,v_size,num_heads,embedding_size)
        self.mh2 = MultiHead(q_size,k_size,v_size,num_heads,embedding_size)
        self.ln1 = nn.LayerNorm(embedding_size)
        self.f1 = FFN(embedding_size,hidden_size,embedding_size)
        self.ln2 = nn.LayerNorm(embedding_size)
        self.ln3 = nn.LayerNorm(embedding_size)
    
    def forward(self,x,state):
        x = x + self.mh1(x,x,x)
        x = self.ln1(x)
        K = state
        V = state
        x = x + self.mh2(x,K,V)
        x = self.ln2(x)
        x = x + self.f1(x)
        x = self.ln3(x)
        return x
```

```python
class Decoder(nn.Module):
    
    
    def __init__(self,q_size,k_size,v_size,num_heads,embedding_size,hidden_size,block_size,vocab_size):
        super().__init__()
        self.emdedding = nn.Embedding(vocab_size,embedding_size)
        self.blks = [DecoderBlock(q_size,k_size,v_size,num_heads,embedding_size,hidden_size) for _ in range(block_size)]
        self.dense = nn.Linear(embedding_size,vocab_size)
        self.sigmoid = nn.Sigmoid()
        
    def forward(self,x,state):
        for blk in self.blks:
            x = blk(x,state)
        logits = self.dense(x)
        probs = self.sigmoid(logits)
        return probs
```

```python
class NMT(nn.Module):
    
    def __init__(self,q_size,k_size,v_size,num_heads,embedding_size,hidden_size,block_size,
                 input_vocab_size,output_vocab_size):
        super().__init__()
        self.encoder = Encoder(q_size,k_size,v_size,num_heads,embedding_size,hidden_size,block_size,input_vocab_size)
        self.decoder = Decoder(q_size,k_size,v_size,num_heads,embedding_size,hidden_size,block_size,output_vocab_size)
        self.input_embedding = nn.Embedding(input_vocab_size,embedding_size)
        self.output_embedding = nn.Embedding(output_vocab_size,embedding_size)
        
    
    def forward(self,x1,x2):
        x1 = self.input_embedding(x1)
        x2 = self.output_embedding(x2)
        x1 = self.encoder(x1)
        probs = self.decoder(x2,x1)
        return probs
    
```

```python
ch_char2index = {}
index2ch_char = {}
en_char2index = {}
index2en_char = {}
start_char = '<beg>'
end_char = '<end>'
ch_char2index[start_char] = 0
index2ch_char[0] = start_char
ch_char2index[end_char] = 1
index2ch_char[1] = end_char


en_char2index[start_char] = 0
index2en_char[0] = start_char
en_char2index[end_char] = 1
index2en_char[1] = end_char

for i,c in enumerate(ch_chars):
    ch_char2index[c] = i + 2
    index2ch_char[i+2] = c
for i,c in enumerate(en_chars):
    en_char2index[c] = i + 2
    index2en_char[i+2] = c
```

```python
q_size = 200
k_size = 200
v_size = 200
num_heads = 10
embedding_size = 200
hidden_size = 100
block_size = 5
input_vocab_size = len(ch_chars)+2
output_vocab_size = len(en_chars)+2
```

```python
all_x1 = []
all_x2 = []
all_y = []
```

```python
for ch_str,en_str in zip(ch,en):
    x1 = [ch_char2index[c] for c in ch_str]
    tmp = [en_char2index[c] for c in en_str]
    x2 = [0] + tmp
    y = tmp + [1]
    x1 = torch.tensor(x1).view(1,-1)
    x2 = torch.tensor(x2).view(1,-1)
    y = torch.tensor(y)
    all_x1.append(x1)
    all_x2.append(x2)
    all_y.append(y)
```

```python
op = optim.Adam(nmt.parameters(),lr=0.0001)
```

```python
criterion = nn.CrossEntropyLoss()
```

```python
for k in range(10000):
    avg = 0
    for i in torch.randint(0,len(all_x1),(10,)):
        x1 = all_x1[i]
        x2 = all_x2[i]
        y = all_y[i]
        pred = nmt(x1,x2)
        loss = criterion(pred.view(len(y),-1),y)
        op.zero_grad()
        loss.backward()
        op.step()
        avg += loss
    if k % 100 == 0:
        print(avg / 10)
```
